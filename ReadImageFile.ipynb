```python
from google.colab import drive

# mounting dataset from gdrive
drive.mount('/content/gdrive', force_remount=True)

# dataset path
root_path = 'gdrive/My Drive/Colab Notebooks/Pengolahan_Citra_01/Lab01 Data/'

print("Path root : ", root_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_image(image_path):
    bgr_img = cv2.imread(image_path)
    rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
    return rgb_img

def show_image(rgb_img):
    plt.figure(figsize=(10,20))
    plt.imshow(rgb_img)
    plt.axis('off')
```

```python
jpg_img = read_image(root_path + 'sample jpg.jpg')
show_image(jpg_img)
```

```python
png_img = read_image(root_path + 'sample png.png')
show_image(png_img)
```

```python
bmp_img = read_image(root_path + 'sample bmp.BMP')
show_image(bmp_img)
```

```python
cap = cv2.VideoCapture(root_path + 'sample gif.gif')
ret, bgr_img = cap.read()
if ret:
  gif_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
  show_image(gif_img)
cap.release()
```

```python
ax2 = plt.figure(figsize=(20,10))
plt.subplot(1,4,1)
plt.imshow(jpg_img)
plt.title('JPEG Image')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(png_img)
plt.title('PNG Image')
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(gif_img)
plt.title('GIF Image')
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(bmp_img)
plt.title('BMP Image')
plt.axis('off')
```

```python
# image resolution
jpg_img.shape

# insert the type of pixel in the image
jpg_img[1,1,1].dtype
```

```python
red = jpg_img[:,:,0]
green = jpg_img[:,:,1]
blue = jpg_img[:,:,2]

# show each channel
ax3 = plt.figure(figsize=(20,10))
plt.subplot(1,4,1)
plt.imshow(jpg_img)
plt.title('RGB Channels')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(red, cmap='gray')
plt.title('Red Channels')
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(green, cmap='gray')
plt.title('Green Channels')
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(blue, cmap='gray')
plt.title('Blue Channels')
plt.axis('off')
```
